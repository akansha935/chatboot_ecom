##4) Create Q&A Dataset:

#### I created a custom Question & Answer dataset from the Flipkart product data to train the chatbot. A loop was used to read each product’s details such as product name, description, brand, retail price, and discounted price, and multiple question-answer pairs were generated automatically for every product. After creating the Q&A data, it was converted into a new dataframe, saved as flipkart_qa_dataset.csv, and its shape was displayed to confirm the dataset was created successfully.

```python
qa_data = []

for _, row in df.iterrows():

    product = str(row["product_name"])
    description = str(row["description"])[:250]

    brand = str(row["brand"])
    retail_price = str(row["retail_price"])
    discounted_price = str(row["discounted_price"])

    # Clean format function style answers

    # Q1: Product description
    qa_data.append({
        "text": f"Question: What is {product}?\nAnswer: {product} is a product. {description}"
    })

    # Q2: Brand
    qa_data.append({
        "text": f"Question: Which brand makes {product}?\nAnswer: The brand of this product is {brand}."
    })

    # Q3: Retail price
    qa_data.append({
        "text": f"Question: What is the retail price of {product}?\nAnswer: The retail price is ₹{retail_price}."
    })

    # Q4: Discount price
    qa_data.append({
        "text": f"Question: What is the discounted price of {product}?\nAnswer: The discounted price is ₹{discounted_price}."
    })


# Convert to dataframe
qa_df = pd.DataFrame(qa_data)

print("Q&A Dataset Shape:", qa_df.shape)

# Save dataset
qa_df.to_csv("/content/flipkart_qa_dataset.csv", index=False)

print("Q&A dataset saved successfully!")
```

####I displayed the first 20 rows of the generated Q&A dataset using qa_df.head(20). This step helps verify that the chatbot training data is properly formatted before model training.

##5)  LOAD DISTILGPT2:

####I loaded the pre-trained GPT-2 based model distilgpt2 and its tokenizer using the Hugging Face Transformers library. The tokenizer converts text into tokens that the model can understand, while AutoModelForCausalLM loads the language model for chatbot text generation. I also set tokenizer.pad_token = tokenizer.eos_token because GPT-2 models do not have a default padding token, and this prevents errors during training.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token
```

#####I converted the Q&A dataframe into a Hugging Face dataset format using Dataset.from_pandas(qa_df). This conversion is necessary because the Hugging Face Trainer works efficiently with datasets in this format during tokenization and model training. The new dataset object now contains all chatbot question-answer pairs in a structure compatible with the training pipeline.

```python
#convert to hugging face dataset
dataset = Dataset.from_pandas(qa_df)
```

##6) Tokenization:

####I defined a tokenization function to prepare the dataset for training the language model. The function converts each text entry into token IDs using the tokenizer, applies padding and truncation to ensure all inputs are of equal length (64 tokens), and then creates labels by copying the input IDs for causal language modeling.

```python
def tokenize_function(example):

    tokens = tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=64
    )

    # Labels for language modeling
    tokens["labels"] = tokens["input_ids"].copy()

    return tokens


tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

##7) Training Arguments:

#####In this step, I configured and started training the chatbot model using the Hugging Face Trainer API. The TrainingArguments define how the model will be trained, including 3 epochs, batch size of 8, mixed precision (fp16=True) for faster GPU performance, and reduced logging and checkpoint saving to make training more efficient.  
#####Then, the Trainer was initialized with the model, training settings, and tokenized dataset, and training was started using trainer.train().

```python
training_args = TrainingArguments(

    output_dir="./results",

    # Faster training
    num_train_epochs=3,

    # Batch size
    per_device_train_batch_size=8,

    # Save less often
    save_steps=2000,

    # Log less often
    logging_steps=500,

     # Use mixed precision
    fp16=True,

    # Faster loading
    dataloader_num_workers=2,

    # Keep only 1 checkpoint
    save_total_limit=1,

    # Disable reports
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset
)

trainer.train()
```

#####The output shows the training progress across 4500 steps, where the loss steadily decreases from around 1.21 to 0.49, indicating that the model is learning well. The message ForCausalLMLoss means the model is using the correct default loss function for text generation. Finally, the model was successfully saved after training completion.

##8)  SAVE MODEL:

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_path = "/content/flipkart_chatbot_model"

tokenizer = AutoTokenizer.from_pretrained(model_path, local_files_only=True)
model = AutoModelForCausalLM.from_pretrained(model_path, local_files_only=True)
```

```python
import os

model_directory = "/content/flipkart_chatbot_model"

if os.path.exists(model_directory):
    print(f"The model directory '{model_directory}' exists.")
else:
    print(f"The model directory '{model_directory}' does NOT exist.")
```

####I created a function that takes user input, converts it into tokens, and passes it to the trained model to generate a response. The output is then decoded back into readable text and returned as the chatbot’s answer.

```python
def qa_chat(user_input, history):

    prompt = f"Question: {user_input}\nAnswer:"

    inputs = tokenizer(prompt, return_tensors="pt")

    output = model.generate(
        **inputs,
        max_new_tokens=50,
        temperature=0.2,
        do_sample=False,
        pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(output[0], skip_special_tokens=True)

    answer = response.split("Answer:")[-1].strip()

    return answer
```

####I built the chatbot interface using Gradio. The ChatInterface connects my qa_chat function to a simple web UI where users can type questions and get answers instantly. Finally, chatbot.launch() starts the application and opens the chatbot in a browser for interaction.

```python
import gradio as gr

chatbot = gr.ChatInterface(
    fn=qa_chat,
    title="Q&A Chatbot",
    description="Ask questions about products and get answers"
)

chatbot.launch()
```
